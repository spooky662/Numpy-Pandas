```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('Salary_Data.csv')
```

```python
salarios = dados['Salary'].to_numpy()
temposervico = dados['YearsExperience'].to_numpy()
```

```python
mediasalarios = np.mean(salarios)
print("Média de Salarios: ", mediasalarios)
```

```python
mediatemposervico = np.mean(temposervico)
print("Média de Tempo de Serviço: ", mediatemposervico)
```

```python
salarios5anos = salarios[temposervico > 5]
mediasalarios5anos = np.mean(salarios5anos)
print("Salário médio dos funcionários com mais de 5 anos de serviço: ", mediasalarios5anos)
```

```python
plt.bar(temposervico, salarios)
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário')
plt.title('Salários x Tempo de Serviço')
plt.show()
```

```python
salariosaumento = salarios * 1.10
print("Salários após aumento de 10%: ", salariosaumento)
```
