```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix = pd.read_csv('Netflix TV Shows and Movies.csv')
```

```python
netflix.head(10)
```

```python
netflix.tail(10)
```

```python
# Filme com o maior imdb_score
melhor_filme = netflix.loc[netflix['imdb_score'].idxmax()]

# Filme com o pior imdb_score
pior_filme = netflix.loc[netflix['imdb_score'].idxmin()]

melhor_filme, pior_filme
```

```python
duracaoserie = netflix[netflix['type'] == 'SHOW']['runtime'].to_numpy()
duracaofilmes = netflix[netflix['type'] == 'MOVIE']['runtime'].to_numpy()

mediaduracaoserie = np.mean(duracaoserie)
mediaduracaofilme = np.mean(duracaofilmes)

mediaduracaoserie, mediaduracaofilme
```

```python
seriemaiorduracao = np.max(duracaoserie)

filmemaiorduracao = np.max(duracaofilmes)

seriemaiorduracao, filmemaiorduracao
```

```python
filmes_por_ano = netflix[netflix['type'] == 'Movie'].groupby('release_year').size()

ano_max_filmes = filmes_por_ano.max()
quantidade_max_filmes = filmes_por_ano.max()

ano_max_filmes, quantidade_max_filmes
```

```python
categorias_age_certification = netflix['age_certification'].value_counts()

categorias_age_certification
```

```python
campos_vazios = netflix.isnull().sum()

colunas_vazias = campos_vazios[campos_vazios > 0]

colunas_vazias
```

```python
media_scores_por_ano = netflix.groupby(['release_year', 'type'])['imdb_score'].mean().unstack()

# Plotar gráfico
media_scores_por_ano.plot(kind='bar', figsize=(12, 6))
plt.title('Média de IMDB Scores ao Longo dos Anos')
plt.xlabel('Ano de Lançamento')
plt.ylabel('Média de IMDB Score')
plt.legend(title='Tipo')
plt.show()
```

```python
votantes_por_tipo = netflix.groupby('type')['imdb_votes'].sum()

votantes_por_tipo.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Proporção de Votantes para Shows e Filmes')
plt.ylabel('')
plt.show()
```
